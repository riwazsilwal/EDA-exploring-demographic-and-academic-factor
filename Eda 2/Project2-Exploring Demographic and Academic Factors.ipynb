## lets see first the no of male and female student using histogram.

```python
sns.histplot(df['sex'],kde=False,
             bins=6,color='blue',
             )
#add labels
plt.title('Age count')
plt.xlabel('sex')
plt.ylabel('count')
```

* So the above histogram depicts that the female students are more than that of male student.
* Also female are the older ones.

```python
sns.barplot(data=df, x="age",y="family_relationship")
```

* from the above barchart between the student of different age with thier family relaton ship,
* As the age is increasing, the student are more responsible and has good family relationship.
* Age group of 16 doesnt seem to have good relationship with thier family as compared to other age groups.

```python
import seaborn as sns
sns.relplot(data=df, x="age",y="absences",kind="line")
```

* from the above line graph, 
* The age group of 15 are less absent as compared to others.
* As the age increases , the absent rate slightly increases and is highest at age group of 22.

```python
sns.scatterplot(x=df['free_time'],y=df['age'])
plt.title('scatterplot of Absecnces vs age')
plt.xlabel('free_time')
plt.ylabel('age')
```

* Here from the scatterplot between absences and age .
* As the age increases, the free time is also increasing .
* Here we have an outlier and it is age group of 22.

```python
# A histogram depiciting a relationship between age group and parents status.
sns.histplot(df['parent_status'],kde=False,
             bins=6,color='blue',
             )
#add labels
plt.title('Age count')
plt.xlabel('parent_status')
plt.ylabel('count')
```

* The above histogram is negatively skewed ,
* The parents of student that are living together are more in number than living apart.

```python
# A boxplot of age .
sns.boxplot(data=df,x="health",hue="sex")
```

* The boxplot is normal distribution since medain is likely to be inbetween.
* It has no outlier and is dispersed from 3.0 to 5.0.
* The median is 4.

```python
# A scatter plot of age and class failures
sns.scatterplot(x=df['class_failures'],y=df['age'],hue=df['sex'])
plt.title('scatterplot of class_failures vs age')
plt.xlabel('class_failures')
plt.ylabel('age')
plt.show()
```

* Here in the above scatterplot between class_failures and age ,
* we come to know are age group 0f 21 and 22 female are more failed.
* Here age group 22 and 21 are outliers.

```python
corrmat=df.corr()
f,ax=plt.subplots(figsize=(16,12))
sns.heatmap(corrmat,vmax=.8,square=True);
```

* There is a positive correlation between age and free time. As age increases, free time also increases.
* The correlation heatmap provides an overview of the relationships between different variables in the dataset.
* It helps identify potential patterns and dependencies.
